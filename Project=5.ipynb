# Define train and test

```python
Atrain=A[0:500]
Atest=A[500:577]
```

```python
import seaborn as sb
sb.distplot(Atrain.Close)
```

```python
A1train=Atrain[["Close"]].values
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
Atrain_std=ss.fit_transform(A1train)
```

```python
sb.distplot(Atrain_std)
```

```python
Atrain.shape
```

```python
X=[]
Y=[]
```

```python
for i in range(5,500,1):
    X.append(A1train[i-5:i])
    Y.append(A1train[i])
```

```python
import numpy as np
X_arr=np.array(X)
Y_arr=np.array(Y)
```

```python
X_arr.shape
```

# ANN

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout,LSTM

nn=Sequential()
nn.add(LSTM(units=100,return_sequences=True,input_shape=(5,1)))
nn.add(Dropout(0.2))
nn.add(LSTM(units=100))
nn.add(Dropout(0.2))
nn.add(Dense(1))

nn.compile(optimizer="adam",loss="mean_squared_error")
nn.fit(X_arr,Y_arr,epochs=500)
```

# Test

```python
Atest.head()
```

```python
A1test=Atest[["Close"]]
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
Atest_std=ss.fit_transform(A1test)
```

```python
Atest_std.shape
```

```python
Xtest=[]
Ytest=[]
```

```python
for i in range(5,76,1):
    Xtest.append(Atest_std[i-5:i])
    Ytest.append(Atest_std[i])
```

```python
Xt_arr=np.array(Xtest)
Yt_arr=np.array(Ytest)
```

```python
pred_std=nn.predict(Xt_arr)
```

```python
pred=ss.inverse_transform(pred_std)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(Ytest,pred)
```
