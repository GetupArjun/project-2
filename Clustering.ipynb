```python
B=A[["MonthlyIncome", "HourlyRate"]]
```

```python
B.head()
```

```python
from sklearn.cluster import KMeans
K=list(range(1,15,1))
wcss=[]
for i in range(1,15,1):
        km=KMeans(n_clusters=i)
        model=km.fit(B)
        wcss.append(model.inertia_)    
```

```python
wcss
```

```python
list(zip(K,wcss))
```

```python
import matplotlib.pyplot as plt
plt.scatter(K,wcss,c="red")
plt.plot(K,wcss,c="blue")
plt.xlabel("Number of clusters - K")
plt.ylabel("Within Clusters Sum of Squares")
plt.title("Elbow Curve")
plt.xticks(range(1,15,1))
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=4)
model=km.fit(B)
```

```python
B[["clusters"]]=model.labels_
```

```python
B.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(B["MonthlyIncome"],B["HourlyRate"],c=B["clusters"])
```

```python
B[["JobRole"]]=A.JobRole
```

```python
B.head(20)
```

```python
a = input("Enter any car Make: ")

clusters_no_for_car = B[B.JobRole == a].clusters.values[0]

print("Cars that are similar to your cars are: ")

print(B[B.clusters == clusters_no_for_car].JobRole.values)
```

# Hierarichal Clustering

```python
C=A[['PercentSalaryHike' , 'MonthlyIncome', "JobSatisfaction", "DistanceFromHome"]]
```

```python
from sklearn.cluster import AgglomerativeClustering
agc=AgglomerativeClustering(n_clusters=4)
model=agc.fit(C)
```

```python
model.labels_
```

```python
A["cluster"]=model.labels_
```

```python
A.head(20)
```

```python
A[A.cluster==2][["JobRole"]]
```

```python
from scipy.spatial import distance_matrix
from scipy.cluster.hierarchy import dendrogram, linkage
import pandas as pd
import matplotlib.pyplot as plt
```

```python
D = pd.DataFrame(distance_matrix(C.values,C.values))
dendrogram(linkage(D))
plt.show()
```
