```python
import pandas as pd
```

```python
A=pd.read_csv("C:/Users/Anuj.kumar/Desktop/DATA/Project2/training_set.csv")
pd.set_option("display.max_columns",None)
pd.set_option("display.max_row",None)
```

```python
A.head()
```

# Check missing data

```python
A.isna().sum()
```

# Replace missing values

```python
for i in A.columns:
    if(A[i].dtype=="object"):
        replace=A[i].mode()[0]
        A[i]=A[i].fillna(replace)
    else:
        replace=A[i].mean()
        A[i]=A[i].fillna(replace)
```

```python
A.isna().sum()
```

# Define X and Y

```python
Y=A[["Loan_Status"]]
```

```python
X=A.drop(labels=["Loan_Status","Loan_ID"],axis=1)
```

```python
cat=[]
con=[]
```

```python
for i in X.columns:
    if(X[i].dtype=="object"):
        cat.append(i)
    else:
        con.append(i)
```

# Preprocessing

```python
Xa=pd.get_dummies(X[cat])
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
Xb=pd.DataFrame(ss.fit_transform(X[con]),columns=con)
X1=Xb.join(Xa)
```

```python
X1.head()
```

# Divide Data into training and testing set

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X1,Y,test_size=0.2,random_state=26)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
model = lr.fit(xtrain,ytrain)

from sklearn.metrics import confusion_matrix,accuracy_score

pred_ts=model.predict(xtest)
pred_tr=model.predict(xtrain)

training_acc=accuracy_score(ytrain,pred_tr)
print("Training_acc: ",round(training_acc,2))
print(confusion_matrix(ytrain,pred_tr))

testing_acc=accuracy_score(ytest,pred_ts)
print("Testing_acc: ",round(testing_acc,2))
print(confusion_matrix(ytest,pred_ts))
```

# Anova Creation

```python
def anova(df,con,cat):
    from statsmodels.formula.api import ols
    from statsmodels.stats.anova import anova_lm
    rel=con+ "~"+cat
    model=ols(rel,df).fit()
    anova_result=anova_lm(model)
    Df=pd.DataFrame(anova_result)
    q=Df["PR(>F)"][cat]
    q=round(q,4)
    return q
```
