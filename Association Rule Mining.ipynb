```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
import pandas as pd
A = pd.DataFrame(te.fit_transform(A),columns=te.columns_)
```

```python
A
```

```python
from mlxtend.frequent_patterns import apriori
C = apriori(A,use_colnames=True)
```

```python
C
```

```python
Q=[]
for i in C.itemsets:
    Q.append(len(i))
```

```python
C["Length"]=Q
```

```python
CT=[]
for i in A.columns:
    CT.append(A[A[i]==True][i].count())
```

```python
CT
```

```python
ind={}
x=0
for i in A.columns:
    ind[i]=CT[x]
    x=x+1
```

```python
ind
```

```python
Q1=[]
Q2=[]
Q3=[]
for i in A.columns:
    for j in A.columns:
        Q1.append(i)
        Q2.append(j)
        Q3.append(A[(A[i]==True) & (A[j]==True)].shape[0])
```

```python
A3=pd.DataFrame([Q1,Q2,Q3]).T
```

```python
A3.columns=["item1","item2","Paircount"]
```

```python
A3
```

```python
W1=[]
for i in A3.item1:
    W1.append(ind[i])
```

```python
A3["individual"]=W1
```

```python
A3
```

```python
A3["Confidence"]=A3.Paircount/A3.individual
```
